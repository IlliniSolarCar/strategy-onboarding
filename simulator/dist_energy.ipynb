```python
import glob
import pandas as pd
import re
import matplotlib.pyplot as plt
import numpy as np

%matplotlib widget

p = '[\d]+'
dists = []
energies = []
avg_targets = []

files = glob.glob("logs/*.csv")
for file in files:

    df = pd.read_csv(file)
    avg_target = np.mean(df['target_mph'])
    avg_targets.append(avg_target)

    for num in re.finditer(p, file):
        if(len(dists) == len(energies)):
            dists.append(int(num[0]))
        else:
            energies.append(int(num[0]))

plt.scatter(dists, energies, c=avg_targets)
plt.xlabel('miles earned')
plt.ylabel('watthours left')
plt.colorbar(label='avg target mph')


```
